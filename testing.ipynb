```python
import pandas as pd
import numpy as np
import pickle
import tensorflow as tf
```

```python
project_path='/content/drive/MyDrive/Github/ANN_classification/'
```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}

```

```python
test_df = pd.DataFrame(input_data, index=[0])
test_df.head()
```

### Load all transorming models

```python
# label Encoder
labbel_encoder_gender=pickle.load(open(project_path+'label_encoder_gender.pkl','rb'))
one_hot_encoder_geo=pickle.load(open(project_path+'one_hot_encoder_geo.pkl','rb'))
standard=pickle.load(open(project_path+'standard.pkl','rb'))
```

```python
test_df['Gender']=labbel_encoder_gender.transform(test_df['Gender'])
geo_encoded = one_hot_encoder_geo.transform([test_df['Geography']]).toarray()
one_hot_geo_df=pd.DataFrame(geo_encoded,columns=one_hot_encoder_geo.get_feature_names_out(['Geography']))
test_df=pd.concat([test_df.drop(['Geography'],axis=1),one_hot_geo_df],axis=1)
test_df=standard.transform(test_df)

```

```python
test_df
```

```python
#load model
model=tf.keras.models.load_model(project_path+'model.h5')
```

```python
model.summary()
```

```python
predict_value=model.predict(test_df)
```

```python
predict_value
```

```python
if predict_value[0][0]>0.5:
  print('Customer will join')
else:
  print('Customer will not join')
```
